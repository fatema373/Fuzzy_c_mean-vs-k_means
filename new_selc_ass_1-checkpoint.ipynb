```python
def correlated_features(dataset,threshold):
    correlation_matrix = dataset.corr()
    correlated = set()
    for i in range(len(correlation_matrix.columns)):
        for j in range(i):
            if abs(correlation_matrix.iloc[i,j])>threshold:
                correlated.add(correlation_matrix.columns[i])
    return correlated
```

```python
correlated=correlated_features(df,0.8)
```

```python
df.drop(correlated,axis=1,inplace=True)
```

# Fuzzy C_Means algorithm

# Initialization 

```python
def initiaizing(n,c):
    weights_matrix = np.random.random((n,c))
    for i in range(n):
        randomly = weights_matrix[i]
        summ = np.sum(randomly)
        for k in range(c):
            weights_matrix[i][k]/=summ    
    return weights_matrix
```

```python
we = initiaizing(df.shape[0],2)
we.shape
we
```

# Compute centers

```python
def compute_centroids(weights,index,data,m):
    rows = data.shape[0]
    cols = data.shape[1]
    centroid_list = np.random.random(cols)
    for i in range (cols):
        nom=0
        dom=0
        for k in range (rows):
            nom+=((weights[k][index])**m)*data.iloc[k,i]
            dom+=(weights[k][index])**m
        centroid_list[i]=(nom/dom)
    return centroid_list
```

```python
compute_centroids(we,0,df,2)
```

```python
def euclidean_distance (p,c):
    summ=0
    for i in range (len(c)):
        summ+=(p[i]-c[i])**2
    return np.sqrt(summ)
```

```python
# calculate distance between points and centers
def calculate_dist(c,data):
    rows = data.shape[0]
    cols = data.shape[1]
    dist = []
    for i in range (rows):
        point = []
        for j in range (len(c)):
            point.append(euclidean_distance(c[j],data.iloc[i]))
        dist.append(point)
    return dist
```

# updating weights

```python
def update_Weight(weight,data,C,m):
    dom = []
    dist=calculate_dist(C,data)
    for i in range(len(data)):
        summ=0
        for j in range (2):
            summ += (1/dist[i][j])**(1/(m-1))
        dom.append(summ)
        
    for i in range(len(data)):
        for j in range (2):
            weight[i][j]= (1/dist[i][j])**(1/(m-1))/dom[i]    
    return weight
```

```python
def C_means_Algorithm(c,data,m,max_iterations):
    weight= np.asarray(initiaizing(df.shape[0],c))
    center = []
    for i in range(max_iterations):
        temp=[]
        for j in range (c):
#             print(0)
            temp.append(compute_centroids(weight,j,data,m))
            print(temp)
        new_metric=np.asarray(update_Weight(weight,data,temp,m))
        center=temp
#         if(np.max(abs(new_metric-weight))<E):
#             break
        weight=new_metric
        
    return (weight,center)
```

# SVM

```python
from sklearn import svm
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df,occupancy)
```

```python
cls = svm.SVC(kernel = 'poly')
cls.fit(x_train,y_train)
pred = cls.predict(x_test)
print('accuracy:',metrics.accuracy_score(y_test,y_pred=pred))
print('precesion:',metrics.recall_score(y_test,y_pred=pred))
print('recall:',metrics.precision_score(y_test,y_pred=pred))
print(metrics.classification_report(y_test,y_pred=pred))
```
